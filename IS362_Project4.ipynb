```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import linear_model
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/mushroom/agaricus-lepiota.data',sep=',', 
                         header=None, 
                         usecols=[0,3,5,9], 
                         names=["Edible","Cap Color","Odor","Gill Color"])

df.replace(to_replace={"Edible":{'p': 1, 'e': 0}}, inplace=True)
df.replace(to_replace={"Cap Color":{'n':0, 'b':1, 'c':2, 'g':3, 'r':4, 'p':5, 'u':6, 'e':7, 'w':8, 'y':9}}, inplace=True)
df.replace(to_replace={"Odor":{'a':0, 'l':1, 'c':2, 'y':3, 'f':4, 'm':5, 'n':6, 'p':7, 's':8}}, inplace=True)
df.replace(to_replace={"Gill Color":{'k':0, 'n':1, 'b':2, 'h':3, 'g':4, 'r':5, 'o':6, 'p':7, 'u':8, 'e':9, 'w':10, 'y':11}}, inplace=True)

df.head()
```

## Determining total number of edible mushroom and  poisonous.

```python
df.groupby('Edible').count()
```

Total number of edible mushrooms in this database is 4208, and 3916 of them are poisonous. 

```python
df.groupby(['Edible', 'Odor']).count()
```

There are 3 edible mushrooms, odors 0, 1, 6. Also, odor 6 is shown in the poisonous mushrooms.

## Deterimining mushrooms edibility by Cap Color.

```python
df.groupby(['Edible', 'Cap Color']).count()
```

Mushrooms with a Cap Color of 4 or 6 are edible, and no mushroom with these colors is shown on poisonous category.

## Deterimining mushrooms edibility by Gill Color.

```python
df.groupby(['Edible', 'Gill Color']).count()
```

Mushrooms with a Gill Color of 6 and 9 always be edible.
Mushrooms with a Gill Color of 3 and 5 always be poisonous.

Based on the finding the "odor" category will be the most reliable method of predicting mushrooms edibility.
